```python
import logging
logging.basicConfig(format='%(asctime)s: %(levelname)s: %(message)s', level=logging.INFO)
```

```python
def get_description_from_mysql(sql_database, sql_table, column_name):
    import mysql.connector
    sql_db = mysql.connector.connect(
        host = '127.0.0.1',
        user = 'root',
        password = '',
        database = sql_database
    )
    cursor = sql_db.cursor()

    # to check whether TABLE EXISTS OR NOT
    cursor.execute("SHOW TABLES LIKE '{}'".format(sql_table))
    temp_result = cursor.fetchone()
    if (temp_result):
        logging.info("TABLE {} exists in DATABASE {}".format(sql_table, sql_database))
    else:
        logging.info("TABLE {} exists in DATABASE {}".format(sql_table, sql_database))
        
    sql_select = "SELECT {}, {} FROM {}".format(column_name[0], column_name[1], sql_table)
    cursor.execute(sql_select)
    result = cursor.fetchall()
    
    return result

sql_database = "crowdfunding_db"
sql_table = "v3_latest_zeczec_basic_project_information"
column_name = ['project_name', 'description_text']
p_description = get_description_from_mysql(sql_database, sql_table, column_name)
print(len(p_description))
```

```python
# to include only Chinese (excluding punctuation, number, and alphabet)
def is_chinese(uchar):
    if uchar >= u'\u4E00' and uchar <= u'\u9fa5':
        return True
    else:
        return False
# def is_punctuation(uchar):
#     if ()

def format_string(content):
    content_str = ""
    for i in content:
        if is_chinese(i):
            content_str += i
        else:
            content_str += " " # to add space when the characteristic is not Chinese
    content_str = content_str.strip()
    content_str = content_str.replace("  ", " ")
    return content_str
```

```python
# store the description as corpus -> list
output_filename = "zeczec_description_text.txt"
output_f = open(output_filename, 'w', encoding='utf-8')
text_num = 0

corpus = []
for p in p_description:
    temp_text = p[1]
    temp_text = format_string(temp_text)
    output_f.write(temp_text+'\n')
    
    description_list = [p[0], temp_text]
    corpus.append(description_list)
    
    text_num+=1
    if text_num%100 == 0:
        logging.info("Processed: {} description".format(str(text_num)))
output_f.close()
print(corpus[:3])
```

```python
from ckiptagger import WS, POS

ws = WS("./data")
pos = POS("./data")
pos_stopword = ['Caa', 'Cab', 'Cba','D', 'Di', 'Da', 'Dk', 'DE', 'WHITESPACE', 'T', 'SHI', 'P', 'FW', 'V_2']

output_cut_file = open('zeczec_description_cut.txt', 'w', encoding='utf-8')
text_num = 0

corpus_cut = []
logging.info("start cutting words")
for p in corpus:
    ws_result = ws([p[1]])
    pos_result = pos(ws_result)
    
    ws_value = ws_result[0]
    pos_value = pos_result[0]
    result = zip(ws_value, pos_value)
    
    cut_description = ""
    for word in result:
        word_list = list(word)
        if word_list[1] not in pos_stopword:
            cut_description += word_list[0]+" "
            output_cut_file.write(word_list[0]+" ")
    output_cut_file.write('\n')
    temp_list = [p[0], p[1], cut_description]
    corpus_cut.append(temp_list)
    
    text_num += 1
    if text_num%100 == 0:
        logging.info("Cut {} row description".format(str(text_num)))
        output_cut_file.flush()
output_cut_file.close()
logging.info("Cut all description")
print(corpus_cut[:3])
```

```python
# update the cut description into the table
def update_cut_mysql(sql_database, sql_table, corpus_cut):
    import mysql.connector
    sql_db = mysql.connector.connect(
        host = '127.0.0.1',
        user = 'root',
        password = '',
        database = sql_database
    )
    cursor = sql_db.cursor()

    # to check whether TABLE EXISTS OR NOT
    cursor.execute("SHOW TABLES LIKE '{}'".format(sql_table))
    temp_result = cursor.fetchone()
    if (temp_result):
        logging.info("TABLE {} exists in DATABASE {}".format(sql_table, sql_database))
    else:
        logging.info("TABLE {} exists in DATABASE {}".format(sql_table, sql_database))
    
    text_num = 0
    for p in corpus_cut:
        sql_update = "UPDATE {} SET description_text_cut='{}' WHERE project_name='{}'".format(sql_table, p[2], p[0])
        cursor.execute(sql_update)
        sql_db.commit()
        text_num += 1
        
        if text_num%100 == 0:
            logging.info("Updated {} rows".format(str(text_num)))
            
    logging.info("Finish updating description_text_cut in {}".format(sql_table))
    
sql_database = "crowdfunding_db"
sql_table = "v3_latest_zeczec_basic_project_information"
update_cut_mysql(sql_database, sql_table, corpus_cut)
```

# create WordCloud without stopword

```python
# clean the stopword from flingv_description_cut.txt
def load_stopwords(filename):
    f = open(filename, "r", encoding='UTF-8')
    lines = f.readlines()
    for idx, val in enumerate(lines):
        lines[idx] = lines[idx].strip('\n')
    return lines
stopword = load_stopwords('tc_stopwords.txt')
print(stopword)
```

```python
# read & import cut text
def load_cuttext(filename):
    f = open(filename, 'r', encoding='UTF-8')
    lines = f.readlines()
    cut_word_list = []
    for idx, val in enumerate(lines):
        words=""
        for c in lines[idx]:
            if c.isspace():
                cut_word_list.append(words)
                words=""
            else:
                words+=c
    return cut_word_list
cut_word_list = load_cuttext('zeczec_description_cut.txt')
print(cut_word_list[:100])
```

```python
no_stopword_list=[]
print("beofore {}".format(str(len(cut_word_list))))
for w in cut_word_list:
    if (len(w)>=2) and (w not in stopword):
        no_stopword_list.append(w)
print("after {}".format(str(len(no_stopword_list))))
print(no_stopword_list[:100])
```

```python
no_stopword_str=""
for w in no_stopword_list:
    no_stopword_str+=(w+" ")
# generate WordCloud
from wordcloud import WordCloud
# set the configuration
font = "/fonts/msjh.ttc"
wc = WordCloud(background_color = "white",
               random_state=42,
              font_path = font)
wc.generate(no_stopword_str)
wc.to_file('zeczec_wordcloud_v2.png')
```

```python
# get the frequency of the word
import nltk
word_fd = nltk.FreqDist(no_stopword_list)
word_fd
```

```python
word_fd_dict = dict(word_fd)
word_fd_dict
```
